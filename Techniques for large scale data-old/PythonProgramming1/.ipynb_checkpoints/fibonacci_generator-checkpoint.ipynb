```python
def fib(limit):
    i = 0
    a = 0
    b = 1
    while i<limit:
        yield b
        b = a+b
        a = b-a
        i += 1
```

```python
def list_fib(limit):
    # Construct a list of Fibonacci series
    fibonacci_series = []
    for i in fib(limit):
        fibonacci_series.append(i)
    return fibonacci_series
```

```python
miau = list_fib(10)
miau
```
